```python
with open("../../output/bdd_data.txt", "r") as f:
    lines = f.readlines()
```

```python
def getstuff(header):
    stuff = []
    for line in lines:
        if line.startswith(header):
            stout = line.split(":")[1].strip()
            try:
                stuff.append(int(stout))
            except ValueError:
                stuff.append(float(stout))
    return stuff
```

```python
dagsizes = getstuff("Final Size")
peak_dagsizes = getstuff("Intermediate Size")
```

```python
xlabel("Maximum Edit Distance")
xlim(1,15)
yscale("log")
ylabel("BDD Size (# nodes)")
ax = gca()
ds, = ax.plot(range(1,16), dagsizes)
pds, = ax.plot(range(1,16), peak_dagsizes)
ds.set_label("DAG Size")
pds.set_label("Peak DAG Size")
ax.legend()
title("Memory Usage as a function of Edit Distance")
```

```python
solns = getstuff("Total solns")
csolns = getstuff("Completable solns")
rate = getstuff("Ratio")
```

```python
xlabel("Maximum Edit Distance")
xlim(1,15)
yscale("log")
ylabel("Completability Rate")
plot(range(1,16), rate)
title("Completability Rate as a function of Edit Distance")
```

```python
sizeratio = [solns[i] / dagsizes[i] for i in range(15)]
```
